```python
import pandas as pd
```

```python
df = pd.read_excel('DataBase.xlsx')
df
```

```python
keywords = [
    "IMU", "IHM", "MRU", "sensor", "aceler", "girosc", "magnet", "angul", "ângul",
    "inercial", "inercial", "filtro", "Kalman", "ruído", "compensação", "calibração",
    "correção", "computação", "monitoramento", "monitor", "moviment", "estabilid",
    "estabiliz", "referência", "Motion Reference Unit", "aquisição", "dados",
    "posição", "velocidade", "aceleração", "oscilaç",
]
```

```python
def getTarget(label):
    for keyword in keywords:
        keyword = keyword.lower().strip()
        label   = label.lower().strip()

        if keyword in label:
            return True
        
    return False

df['target'] = df.label.apply(getTarget)
df
```

```python
target = df.loc[df.target]
target
```
